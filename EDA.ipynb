```python
data=pd.read_csv('/Users/kruthi/Desktop/Han Assignment/Amazon_Customer_Behavior_Survey (1).csv')
print(data.columns)
print(data.info())
print(data.isna().sum())
print(data.describe())
```

```python
#filters the original dataset into four separate groups based on gender categories and then prints these subsets

male = data[data['Gender']=='Male']
female = data[data['Gender']=='Female']
others = data[data['Gender']=='Others']
prefer_not_say = data[data['Gender']=='Prefer not to say']

print(male)
print(female)
print(others)
print(prefer_not_say) 
```

```python
# for visualisations
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
# Filter to only include numeric columns for correlation
numeric_male = male.select_dtypes(include=[np.number])

for i in male.columns.values:
    index=male[i].value_counts().index.values
    val=male[i].value_counts().values
    if (len(index) and len(val)) <=10:
        plt.pie(val,labels=index,autopct='%1.1f%%')
        plt.title(f'the values and their counts related to male category  for  {i} column')
        plt.legend()
        plt.show()
plt.figure(figsize=(17, 6))

# Filter to only include numeric columns for correlation
numeric_male = male.select_dtypes(include=[np.number])
# Calculate correlation matrix
corr = numeric_male.corr(method='spearman')

my_m = np.triu(corr)
sn.heatmap(corr, mask=my_m, annot=True, cmap="Set2")
plt.show()




for i in male.select_dtypes(include='number').columns.values:
    for j in male.select_dtypes(include='number').columns.values:
        sn.distplot(male[i], label=f"{i}", color='red')
        sn.distplot(male[j], label=f"{j}", color="blue")
        plt.title(f"{i} vs {j} in male category")
        plt.legend()
        plt.show()
```
